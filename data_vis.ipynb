```python
import pandas as pd
import matplotlib.pyplot as plt
# Load your dataset (replace 'your_dataset.csv' with your actual dataset file)
df = pd.read_csv(r'C:\Users\alosh\OneDrive\Desktop\VSCODE\DeepthyJ\Dataset\Darknet.csv', delimiter=',')
```

```python
df['Label3'].unique()
```

```python
# Assuming your label column is named 'Label' (replace with your actual column name)
label_counts = df['Label3'].value_counts()
print(label_counts)

# Define custom y-axis points and corresponding labels
custom_y_ticks = [1000, 2000, 3000]  # Set your desired points (in 10000s)
custom_y_tick_labels = ['1000', '2000', '3000']  # Corresponding labels

# Plot the label counts
plt.figure(figsize=(12, 6))
label_counts.plot(kind='bar', color='blue')
plt.title('Features_Dataset_2')
plt.xlabel('Label')
plt.ylabel('Count (in 10000s)')

# Set the custom y-axis points and labels
plt.yticks(custom_y_ticks, custom_y_tick_labels)

plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```
